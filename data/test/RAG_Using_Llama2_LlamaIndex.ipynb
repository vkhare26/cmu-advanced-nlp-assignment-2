```python
# Import the core components from LlamaIndex for building a retrieval-augmented generation (RAG) system
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, ServiceContext

# - VectorStoreIndex: Handles storing and retrieving document embeddings for retrieval-based querying.
# - SimpleDirectoryReader: Reads documents (e.g., PDFs, text files) from a specified directory.
# - ServiceContext: Manages configurations like embedding models and LLM integration.

# Import Hugging Face's LLM integration for LlamaIndex
from llama_index.llms.huggingface import HuggingFaceLLM

# - HuggingFaceLLM: Allows integration of a Hugging Face-based Llama 2 model (or any other Hugging Face-supported model)
#   into LlamaIndex for generating responses.

# Import a prompt template utility from LlamaIndex
from llama_index.core.prompts.prompts import SimpleInputPrompt

# - SimpleInputPrompt: A pre-defined prompt template that formats user inputs before passing them to the LLM.
#   This can be customized to control how queries are structured.
```

```python
# Create a SimpleDirectoryReader instance pointing to the directory containing your data.
# Here, "/content/data" is the directory path where your PDF, text, or other files are stored.
# Then, load_data() reads all the documents from that directory into a list of Document objects.
documents = SimpleDirectoryReader("/data/").load_data()
```

```python
# Define a system prompt that sets the overall behavior of the Q&A assistant.
# The prompt instructs the assistant to provide accurate answers based on the provided context,
# and to explicitly state when it doesn't know the answer rather than fabricating one.
system_prompt = """
You are a Q&A assistant. Your goal is to answer questions as accurately as possible based on the instructions and context provided.
If you don't know the answer, just say that you don't know. Don't try to make up an answer.
"""

# Define a query wrapper prompt using the SimpleInputPrompt class from llama_index.
# This wrapper formats the user's query in a standardized way that the Llama2 model expects.
# Here, the format uses special tokens:
#   <|USER|> to indicate the beginning of the user's message,
#   and <|ASSISTANT|> to indicate where the assistant's response should begin.
# The {query_str} placeholder will be replaced with the actual query text during execution.
query_wrapper_prompt = SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
import torch

# Initialize the HuggingFaceLLM instance with specified configuration.
llm = HuggingFaceLLM(
    # Maximum context length (number of tokens) the model can process.
    context_window=4096,
    
    # Maximum number of new tokens to generate in a response.
    max_new_tokens=256,
    
    # Additional generation parameters passed to the model.
    # Here, temperature is set to 0.0 for deterministic output, and sampling is disabled.
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    
    # The system prompt that guides the overall behavior of the Q&A assistant.
    system_prompt=system_prompt,
    
    # A wrapper prompt that formats user queries in a way that the model expects.
    query_wrapper_prompt=query_wrapper_prompt,
    
    # The name of the tokenizer to use. This should match the model for consistent tokenization.
    tokenizer_name="meta-llama/Llama-2-7b-chat-hf",
    
    # The name of the model to load from Hugging Face. It should be a Llama 2 chat model.
    model_name="meta-llama/Llama-2-7b-chat-hf",
    
    # Automatically choose the appropriate device (e.g., GPU if available, else CPU).
    device_map="auto",
    
    # Additional keyword arguments for model loading:
    # - torch_dtype sets the data type to float16 for reduced memory usage.
    # - load_in_8bit enables loading the model in 8-bit mode for further memory efficiency.
    model_kwargs={"torch_dtype": torch.float16, "load_in_8bit": True}
)
```

```python
# Import HuggingFaceEmbeddings from LangChain.
# This class provides a way to compute text embeddings using pre-trained models hosted on Hugging Face.
from langchain.embeddings import HuggingFaceEmbeddings

# Import LangchainEmbedding from LlamaIndex.
# This wrapper integrates LangChain's embedding capabilities into LlamaIndex, enabling you to use these embeddings 
# when building indexes for retrieval-augmented generation.
from llama_index.embeddings.langchain import LangchainEmbedding

# Import Settings from LlamaIndex core.
# The Settings class is used to configure global or runtime parameters for LlamaIndex, allowing you to customize 
# various aspects of the index behavior or integration settings.
from llama_index.core import Settings
```

```python
# Initialize an embedding model for use in your RAG system.
# This uses LangchainEmbedding as a wrapper to integrate embeddings from LangChain with LlamaIndex.

embed_model = LangchainEmbedding(
    # HuggingFaceEmbeddings is used to generate vector representations (embeddings) of text.
    # We specify a pre-trained model ("sentence-transformers/all-mpnet-base-v2") from Hugging Face,
    # which is known for providing strong sentence-level embeddings.
    HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
)
```

```python
# Import VectorStoreIndex and Settings from llama_index.
# VectorStoreIndex is used to build an index over your documents for efficient retrieval.
# Settings allows you to configure global settings such as the LLM and embedding model.
from llama_index.core import VectorStoreIndex, Settings

# Import SentenceSplitter from the node_parser module.
# SentenceSplitter is a transformation that splits documents into smaller chunks (sentences or paragraphs)
# to better fit within the model's context window.
from llama_index.core.node_parser import SentenceSplitter

# Set the global LLM for LlamaIndex. This LLM is used for generating responses during retrieval-augmented generation.
Settings.llm = llm

# Set the global embedding model for LlamaIndex. 
# Here, 'embed_model' is the instance created earlier (via LangchainEmbedding with HuggingFaceEmbeddings).
Settings.embed_model = embed_model  # Ensure this is named correctly as embed_model

# Create a SentenceSplitter instance.
# The chunk_size parameter determines the maximum token length of each split chunk.
sentence_splitter = SentenceSplitter(chunk_size=1024)

# Build the index using VectorStoreIndex from your documents.
# The from_documents() method processes the list of document objects,
# applies any specified transformations (here, the sentence splitter to break documents into chunks),
# and then creates an index that can later be queried.
index = VectorStoreIndex.from_documents(
    documents,
    transformations=[sentence_splitter]  # Pass the splitter to transform each document into chunks.
)
```

```python
index
```

```python
# Convert the created index into a query engine.
# This query engine provides a higher-level interface for querying the indexed documents.
# When you later pass a query to this engine, it handles the retrieval of relevant document chunks
# and interacts with the LLM to generate an answer based on the retrieved information.
query_engine = index.as_query_engine()
```

```python
# Query the index via the query engine.
# This sends the question "What is Little Italy Days Entertainment Schedule?" to the system.
# The query engine retrieves the most relevant document chunks from the index,
# then uses the configured LLM (e.g., Llama 2) along with the embedding model to generate a response.
response = query_engine.query("What is Little Italy Days Entertainment Schedule?")
```
